```python
with requests.session() as s:
    geturl = 'https://www.linkedin.com/uas/login/'
    posturl = 'https://www.linkedin.com/checkpoint/lg/login-submit/'
    r = s.get(geturl)
    soup = BeautifulSoup(r.content, 'html5lib')
    login_data['csrfToken'] = soup.find('input' , attrs={'name':'csrfToken'})['value']
    login_data['sIdString'] = soup.find('input' , attrs={'name' :'sIdString'})['value']
    login_data['pageInstance'] = soup.find('input' , attrs={'name' :'pageInstance'})['value']
    login_data['trk'] = soup.find('input' , attrs={'name' :'trk'})['value']
    login_data['ac'] = soup.find('input' , attrs={'name' :'ac'})['value']
    login_data['authUUID'] = soup.find('input' , attrs={'name' :'authUUID'})['value']
    login_data['session_redirect'] = soup.find('input' , attrs={'name' :'session_redirect'})['value']
    login_data['loginCsrfParam'] = soup.find('input' , attrs={'name' :'loginCsrfParam'})['value']
    login_data['fp_data'] = soup.find('input' , attrs={'name' :'fp_data'})['value']
    login_data['_d'] = soup.find('input' , attrs={'name' :'_d'})['value']
    login_data['controlId'] = soup.find('input' , attrs={'name' :'controlId'})['value']
    login_data['parentPageKey'] = soup.find('input' , attrs={'name' :'parentPageKey'})['value']
    feedurl = 'https://www.linkedin.com/feed/'
    x = s.post(posturl , data=login_data)
    y = s.get(feedurl)
    usersoup = BeautifulSoup(y.content , 'html5lib')
    print(usersoup.prettify())
```
